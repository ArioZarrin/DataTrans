Import

```python
import os
import sys

from langchain.document_loaders import PyPDFLoader
from langchain.document_loaders import TextLoader
from langchain.document_loaders import Docx2txtLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

from huggingface_hub import notebook_login

import torch
import transformers

from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import pipeline
from langchain import HuggingFacePipeline
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.chat_models import ChatOpenAI

from transformers import TrOCRProcessor, VisionEncoderDecoderModel
import requests
from PIL import Image
```

# OCR

```python
ocr_model = "microsoft/trocr-base-handwritten"

processor1 = TrOCRProcessor.from_pretrained(ocr_model)

model1 = VisionEncoderDecoderModel.from_pretrained(ocr_model)
```

```python
def ocr(image):

  pixel_values = processor1(image, return_tensors="pt").pixel_values

  generated_ids = model1.generate(pixel_values)

  generated_text = processor1.batch_decode(generated_ids, skip_special_tokens=True)[0]

  return generated_text
```

# PDF and Image

```python
document=[]
for file in os.listdir("/content/"):
  if file.endswith(".pdf"):
    pdf_path=file
    loader=PyPDFLoader(pdf_path)
    document.extend(loader.load())
print(len(document))

url = "https://fki.tic.heia-fr.ch/static/img/a01-122-02.jpg"
image = Image.open(requests.get(url, stream=True).raw).convert("RGB")
t = ocr(image)
print(t)
```

# Document into Chunks

```python
document_splitter = CharacterTextSplitter(separator='\n', chunk_size=500, chunk_overlap=100)

document_chunks = document_splitter.split_documents(document)

len(document_chunks)
```

# Sentence-transformers

```python
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

# Setting Up Chroma as our Vector Database

```python
vectordb = Chroma.from_documents(document_chunks, embedding=embeddings, persist_directory='./dataset_documents')

vectordb.persist()
```

```python
model_name = "openlm-research/open_llama_3b_v2"

tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=True)

model = AutoModelForCausalLM.from_pretrained(model_name,
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                              #load_in_8bit=True,
                                              load_in_4bit=True)
```

# Hugging Face Pipeline

```python
pipe=pipeline("text-generation",
              model=model,
              tokenizer=tokenizer,
              torch_dtype=torch.bfloat16,
              device_map='auto',
              max_new_tokens=512,
              min_new_tokens=-1,
              top_k=30
              )
```

```python
llm = HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature': 0})

llm
```

# Memory object

```python
memory = ConversationBufferMemory(memory_key='chat_history', return_messages=True)
```

# Conversation Retrieval QA Chain

```python
qa = ConversationalRetrievalChain.from_llm(llm=llm,
                                             retriever=vectordb.as_retriever(search_kwargs={'k': 6}),
                                             verbose=False, memory=memory)
```

Store result

```python
result = qa({"question":"Please provide the title and a brief overview of the content covered in the document"})

ans1 = result['answer']

result = qa({"question":"Can you give me concise summaries of each chapter or section in the document"})

ans2 = result['answer']

result = qa({"question":"What are the main concepts, terms, or keywords discussed in the document"})

ans3 = result['answer']

result = qa({"question":"Could you extract any illustrative examples, case studies, or real-world scenarios mentioned in the document"})

ans4 = result['answer']

result = qa({"What are the main conclusions, insights, or takeaways that the author presents in the document"})

ans5 = result['answer']

answers = [ans1, ans2, ans3, ans4, ans5]
```
